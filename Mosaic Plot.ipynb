```python
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic
import pandas as pd

data = pd.read_csv('123.csv')
data['Loan_Status_Repaid'] = data['Loan_Status'].replace({'Y': 'Repaid', 'N': 'Not Repaid'})

props = {
    ('Graduate', 'Not Repaid'): {'facecolor': 'lightblue'},
    ('Graduate', 'Repaid'): {'facecolor': 'blue'},
    ('Not Graduate', 'Not Repaid'): {'facecolor': 'lightgray'},
    ('Not Graduate', 'Repaid'): {'facecolor': 'gray'}
}

plt.figure(figsize=(10, 6))
mosaic(data, ['Education', 'Loan_Status_Repaid'], properties=props)
plt.title('Mosaic Plot of Loan Repayment by Education')

plt.subplots_adjust(left=0.2, right=0.9, top=0.9, bottom=0.1)

plt.show()
```
